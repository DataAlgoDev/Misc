## timeit

```python
import timeit
import numpy as np

def example_function(n):
    total = 0
    for i in range(n):
        total += i
    return total



# Measure the time it takes to run example_function with different input sizes
input_sizes = [10, 100, 1000, 10000, 100000]
times = []

for size in input_sizes:
    t = timeit.timeit(lambda: example_function(size), number=100)
    times.append(t)
    print(f"Input size: {size}, Time taken: {t}")

# You can then plot these times using a library like matplotlib to visualize the time complexity
import matplotlib.pyplot as plt

plt.plot(input_sizes, times, marker='o')
plt.xlabel('Input Size')
plt.ylabel('Time Taken (seconds)')
plt.title('Time Complexity Analysis')
plt.show()
```

```python
import timeit
import random

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

# Generate random arrays of different sizes
input_sizes = [10, 100, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]
times = []

for size in input_sizes:
    arr = [random.randint(0, 10000) for _ in range(size)]
    t = timeit.timeit(lambda: bubble_sort(arr), number=1)
    times.append(t)
    print(f"Input size: {size}, Time taken: {t}")

# Plot the results
import matplotlib.pyplot as plt

plt.plot(input_sizes, times, marker='o')
plt.xlabel('Input Size')
plt.ylabel('Time Taken (seconds)')
plt.title('Bubble Sort Time Complexity')
plt.show()
```

```python
import random
import timeit
import matplotlib.pyplot as plt

def bubsort(l1):
    for i in range(len(l1)):
        for j in range(i+1,len(l1)):
            if l1[i] > l1[j]:
                l1[i], l1[j] = l1[j], l1[i]
    return l1

n = [10, 100, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]
t = []

for k in n:
    randomlist = [random.randint(1, k) for _ in range(k)]
    t.append(timeit.timeit(lambda:bubsort(randomlist), number = 1))

plt.plot(n,t, marker='o')
plt.xlabel('Sample size')
plt.ylabel('Time')
plt.title('Time complexity of bubble sort O(N^2)')
plt.show()
```

```python
from random import randint as rand
import matplotlib.pyplot as plt
from typing import List
import timeit

class BasicSorts():
    @staticmethod
    def bubblesort(arr : List[int]) -> None:
        # O(n^2)
        n = len(arr)
        for i in range(n):
            for j in range(i + 1, n):
                if arr[i] > arr[j]:
                    arr[i], arr[j] = arr[j], arr[i]

    @staticmethod
    def selectionsort(arr : List[int]) -> None:
        # O(n^2)
        n = len(arr)
        for i in range(n - 1):
            index = i
            for j in range(i + 1, n):
                if arr[j] < arr[index]:
                    index = j
            if index != i:
                arr[i], arr[index] = arr[index], arr[i]

    @staticmethod
    def insertionsort(my_list : List[int]) -> None:
        # O(n^2)
        for i in range(1, len(my_list)):
            temp = my_list[i]
            j = i - 1
            while temp < my_list[j] and j > -1:
                my_list[j + 1] = my_list[j]
                my_list[j] = temp
                j -= 1
        return my_list

class HeapSort():
    @staticmethod
    # Helper function to maintain heap structure
    def heapfy(arr : List [int], root : int, n : int) -> None:
        new_root = root
        left = 2 * root + 1
        right = 2 * root + 2

        if left < n and arr[left] > arr[new_root]:
            new_root = left

        if right < n and arr[right] > arr[new_root]:
            new_root = right

        if new_root != root:
            arr[new_root], arr[root] = arr[root], arr[new_root]
            HeapSort.heapfy(arr, new_root, n)

    @staticmethod
    # Sorts the array in place
    def heapsort(arr : List[int]) -> None:
        n = len(arr)

        # Builds a Max heap
        for i in range(n // 2 - 1, -1, -1):
            HeapSort.heapfy(arr, i, n)
            
        # Bigger values gets moved to the end and size of heap reduces one at a time. O(n log n) 
        for i in range(n - 1, 0, -1):
            arr[i], arr[0] = arr[0], arr[i]
            HeapSort.heapfy(arr, 0, i)

class MergeSort():
    @staticmethod
    # Helper function that creates new sorted list. O(n)
    def _merge(left : List, right : List[int]) -> List[int]:

        combined = []
        i, j = 0, 0

        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                combined.append(left[i])
                i += 1
            else:
                combined.append(right[j])
                j += 1
        while i < len(left):
            combined.append(left[i])
            i += 1
        while j < len(right):
            combined.append(right[j])
            j += 1
        return combined
    
    @staticmethod
    # O(log n)
    def mergesort(arr : List[int]) -> List[int]:
        # Base case
        if len(arr) <= 1:
            return arr
        mid = int(len(arr) / 2)
        left = MergeSort.mergesort(arr[:mid])
        right = MergeSort.mergesort(arr[mid:])
        return MergeSort._merge(left, right)

if __name__ == "__main__":

    def range_gen(length):
        x = [k for k in range(length)]
        return x

    def random_num_lists(count, length = 100):
        x = range_gen(length)
        arr = []
        temp = []

        for _ in range(count):
            for n in x:
                for _ in range(n):
                    arr.append(rand(0,9))
                temp.append(arr[::])
                arr.clear()
            yield temp[::]
            temp.clear()


    algorithms = {
                    "Bubble Sort"     : BasicSorts.bubblesort,
                    "Selection Sort"  : BasicSorts.selectionsort,
                    "Insertion Sort"  : BasicSorts.insertionsort,
                    "Heap Sort"       : HeapSort.heapsort,
                    "Merge Sort"      : MergeSort.mergesort,
                    "Standard Sort"   : sorted
                  }
    
    colours = ["blue", "orange", "green", "magenta", "purple", "brown"]

    count = len(algorithms)
    length = 1000

    g = list(random_num_lists(count, length))
    y = list()
    r = range_gen(length)

    for i, l in enumerate(g):
        k = list(algorithms.values())
        t = [(timeit.timeit(lambda : k[i](a), globals = globals(), number = 1)) for a in l]
        y.append(t[::])
        t.clear()
    
    plt.figure(figsize=(6, 6))
    plt.title('Plot of n2 sorts')
    plt.xlabel('Array size')
    plt.ylabel('Time taken')

    for i in range(count):
        plt.scatter(r, y[i], label = list(algorithms.keys())[i], color = colours[i], marker="o", s = 2)
        # plt.plot(r, y[i], label = list(algorithms.keys())[i], color = colour[i], linewidth = 1, marker = "o", ms = 2, markerfacecolor='black')

    plt.legend()
    plt.grid(True)
    # plt.show()
```
